```python
import matplotlib.pyplot as plt
import matplotlib.patches as patch
import pandas as pd

#Reading the CSV file
data=pd.read_csv('circle_info.csv')
print(data)

for i,d in data.iterrows():
    #input
    radius=d.radius
    x_center=d.x_center
    y_center=d.y_center
    x1=d.x1
    y1=d.y1
    x2=d.x2
    y2=d.y2

    #calculations
    width=x2-x1
    height=y2-y1
    x_min_limit=min(x1,x2)-3
    x_max_limit=max(x1,x2)+3
    y_min_limit=min(y1,y2)-3
    y_max_limit=max(y1,y2)+3
    hf,ha=plt.subplots(1,1,figsize=(5,5))
    plt.xlim(x_min_limit,x_max_limit)
    plt.ylim(y_min_limit,y_max_limit)
    # plt.xlim(-3,3)
    # plt.ylim(-3,3)
    rect=plt.Rectangle((x1,y1),width,height)
    circle=plt.Circle((x_center,y_center),radius,color="red")
    ha.add_patch(rect)
    ha.add_patch(circle)
plt.show()
```
